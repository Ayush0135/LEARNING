```python
import pandas as pd
```

```python
df=pd.read_excel("/Users/ayush/Desktop/Learning/loan data.xlsx")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Gender'].fillna(df['Gender'].mode()[0],inplace=True)
df['Married'].fillna(df['Married'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
en_data = df[["Gender","Married"]]
en_data
```

```python
pd.get_dummies(en_data)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oeh = OneHotEncoder()
oeh.fit_transform(en_data)
```

```python
oeh.fit_transform(en_data).toarray()
```

```python
ar = oeh.fit_transform(en_data).toarray()
pd.DataFrame(ar,columns = ["Gender_Male", "Gender_Female", "Married_No", "Married_Yes"])
```
